```python
%pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

TOP_MNCs = [
    "Amazon", "Microsoft", "Oracle", "Salesforce", "Google", "Adobe", "ServiceNow"
]

def fetch_linkedin_mnc_jobs(username, password):
    driver = webdriver.Chrome()
    driver.get("https://www.linkedin.com/login")

    driver.find_element(By.ID, "username").send_keys(username)
    driver.find_element(By.ID, "password").send_keys(password)
    driver.find_element(By.XPATH, "//button[@type='submit']").click()
    time.sleep(3)

    search_url = (
        "https://www.linkedin.com/jobs/search/?keywords=ETL%20Developer"
        "&location=Hyderabad%2C%20Telangana%2C%20India&f_TPR=r86400"
    )
    driver.get(search_url)
    time.sleep(5)

    jobs = []
    job_cards = driver.find_elements(By.CLASS_NAME, "job-card-container")

    for card in job_cards:
        try:
            title = card.find_element(By.CLASS_NAME, "job-card-list__title").text
            company = card.find_element(By.CLASS_NAME, "job-card-container__company-name").text
            location = card.find_element(By.CLASS_NAME, "job-card-container__metadata-item").text
            link = card.find_element(By.CLASS_NAME, "job-card-list__title").get_attribute("href")

            # Filter by MNC names
            if any(mnc.lower() in company.lower() for mnc in TOP_MNCs):
                jobs.append({
                    "Title": title,
                    "Company": company,
                    "Location": location,
                    "Link": link
                })
        except Exception as e:
            print("❌ Error parsing card:", e)

    df = pd.DataFrame(jobs)
    df.to_excel("Filtered_LinkedIn_ETL_Jobs_MNCs.xlsx", index=False)
    print(f"✔️ Saved {len(jobs)} MNC jobs to 'Filtered_LinkedIn_ETL_Jobs_MNCs.xlsx'")
    driver.quit()

# Call this with your credentials
fetch_linkedin_mnc_jobs("Ravikalavakuntla.235@gmail.com", "Fr0nt3nD5$")
```
