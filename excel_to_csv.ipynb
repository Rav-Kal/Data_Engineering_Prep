```python
!pip install pandas openpyxl xlrd
import os
import pandas as pd

def convert_excel_to_csv(source_folder, target_folder):
    os.makedirs(target_folder, exist_ok=True)

    # List Excel files
    excel_files = [f for f in os.listdir(source_folder) if f.endswith((".xlsx", ".xls"))]

    if not excel_files:
        print("No Excel files found in source folder.")
        return

    for filename in excel_files:
        excel_path = os.path.join(source_folder, filename)
        try:
            print(f"Reading: {filename}")
            df = pd.read_excel(excel_path)
            base_name = os.path.splitext(filename)[0]
            csv_filename = f"{base_name}.csv"
            csv_path = os.path.join(target_folder, csv_filename)
            df.to_csv(csv_path, index=False)
            print(f"Converted and saved to: {csv_path}")
        except Exception as e:
            print(f"Failed to convert {filename}: {e}")

source_folder = r'D:\Ravi_Data_Engineer\00_Master_Data_Sets\02_excel'
target_folder = r'D:\Ravi_Data_Engineer\00_Master_Data_Sets\01_csv'

convert_excel_to_csv(source_folder, target_folder)
```
